```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/mattharrison/datasets/raw/master/data/2020-jetbrains-python-survey.csv'
```

```python
jb = pd.read_csv(url)
```

```python
jb.head()
```

```python
jb.filter(like='job.role')
```

```python
(jb
    .filter(like=r'job.role.*t')
    .where(jb.isna(), 1)
)
```

```python
(jb
    .filter(like=r'job.role.*t')
    .where(jb.isna(), 1)
    .fillna(0)
)
```

```python
job = (jb
      .filter(like=r'job.role')
      .where(jb.isna(), 1)
      .fillna(0)
      .idxmax(axis='columns')
      .str.replace('job.role', '', regex=False)
      )
```

```python
job
```

```python
dum = pd.get_dummies(job)
```

```python
dum
```

```python
dum.idxmax(axis='columns')
```

```python
i, j = np.where(dum)
```

```python
pd.Series(dum.columns[j], i)
```
