```python
import pandas as pd
import numpy as np
```

```python
series = {
    'index':[0, 1, 2, 3],
    'data':[145, 142, 38, 13],
    'name': 'songs'
}
```

```python
def get(series, idx):
    value_idx = series['index'].index(idx)
    return series['data'][value_idx]
```

```python
get(series, 1)
```

```python
get(series, 2)
```

```python
s2 = pd.Series(['m', 'l', 'xs', 's', 'xl'])
```
